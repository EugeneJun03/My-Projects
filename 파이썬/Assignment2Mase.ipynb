## Assignment2: Maze



- Up   : 0010 = 2
- Down : 1000 = 8
- Right: 0001 = 1
- Left : 0100 = 4

```python
size = int(input())
matrix = [input().split() for i in range(size)]
```

```python
print(matrix)
```

```python
bin_matrix = [['{:04d}'.format(int(bin(int(i))[2:])) for i in row] for row in matrix]
print(bin_matrix)
```

```python
#0,0 -> 2,2
path = [[0,0]]
def direction(p):
    # Down
    if bin_matrix[p[0]][p[1]][0] == "1":
        p[0] += 1
        print(p)
        if p == [size-1,size-1]:
            path.append(p.copy())
            return p
        v = bin_matrix[p[0]][p[1]]
        if v != "0001" and v != "0010" and v != "0100" and v != "1000" and p not in path:
            path.append(p.copy())
            p = direction(p)
            if p == [size-1,size-1]:
                return p
        else:
            print("-------- Dead End")
            p = [p[0]-1,p[1]]
        
    # Left    
    if bin_matrix[p[0]][p[1]][1] == "1":
        p[1] -= 1
        print(p)
        if p == [size-1,size-1]:
            path.append(p.copy())
            return p
        v = bin_matrix[p[0]][p[1]]
        if v != "0001" and v != "0010" and v != "0100" and v != "1000" and p not in path:
            path.append(p.copy())
            p = direction(p)
            if p == [size-1,size-1]:
                return p
        else:
            print("-------- Dead End")
            p = [p[0],p[1]+1]
        
    # Up 
    if bin_matrix[p[0]][p[1]][2] == "1":
        p[0] -= 1
        print(p)
        if p == [size-1,size-1]:
            path.append(p.copy())
            return p
        v = bin_matrix[p[0]][p[1]]
        if v != "0001" and v != "0010" and v != "0100" and v != "1000" and p not in path:
            path.append(p.copy())
            p = direction(p)
            if p == [size-1,size-1]:
                return p
        else:
            print("-------- Dead End")
            p = [p[0]+1,p[1]]
    
    # Right
    if bin_matrix[p[0]][p[1]][3] == "1":
        p[1] += 1
        print(p)
        if p == [size-1,size-1]:
            path.append(p.copy())
            return p
        v = bin_matrix[p[0]][p[1]]
        if v != "0001" and v != "0010" and v != "0100" and v != "1000" and p not in path:
            path.append(p.copy())
            p = direction(p)
            if p == [size-1,size-1]:
                return p
        else:
            print("-------- Dead End")
            p = [p[0],p[1]-1]
    
    # if you should return and restart the rout.
    path.remove(path[-1])
    return path[-1].copy()
```

```python
direction([0,0])
path = [[value + 1 for value in row] for row in path]
print(path)
```
