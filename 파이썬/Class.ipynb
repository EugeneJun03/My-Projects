#### OOP
OOP is a short name of object-oriented programming, which is a programming style based on object(data, method).
Unlike procedure programming this type of programming allow reuse of code thereby increasing productivity and new featuers can be easily buit into the existing code.
- Object : an entity in the real world, a particular instance of a class
- Class a template or blueprint for object

```python
# an simple structure of class
class Coordinate(object):
    def __init__(self, x, y):
        self.x = x
        self.y = y
```

- first you should creat a Class using "class"
- And you need to make a method(fuction) called "__init__"
- "self" is needed to get a value from it's own
- self.VariableName stores a value by instances

- if you call Coordinate several times in different names it gets every different instances

```python
class AutoEmail:
    def __init__(self, name):
        self.name = name
    def __str__(self): # This method is for "print()"
        return str(self.name)

mail_to_park = AutoEmail("PARK")
mail_to_kim = AutoEmail("KIM")

print(mail_to_park)
print(mail_to_kim)
```

##### Class Variables Vs Instance Variables

- Class Variables: it is active in all Instance and subclass

- Instance Variables: changes made to the instance variable affects just on it's variable and typicaly initialized in the constructor of the class.

```python
class AutoEmail2:
    intro = "Hello" # Class Variable

    def __init__(self, name):
        self.name = name # Instance Variable
        
    def say_hello(self):
        return f"{self.intro} {self.name} nice to meet you!"
    
mail_to_park = AutoEmail2("PARK")
print(mail_to_park.say_hello())
```

### Class Resource Control(Public, Protected, Private)
- Public member: available to everyone(default)

- Protected member: accessible from within the class it sub-classes
Add a prefix "_"(single underscore)

- Private member: accessible within the class
Add a prefix "__"(double underscore)

```python
class Student:
    schoolName = "XYZ School" #Public member

    def __init__(self, name, age): # private member
        self.__name = name # privat member
        self._age = age # Protected member
    def display(self):
        print('This is Public member')

std = Student("BILL", 25)
print(std.schoolName)
std.display()
std.__name # private member is not allowed to access outside of the class
```

### Inheritance
Inheritance is like genetic transmission, which passes property to subclass

- child class: It inherits all data from parent class can add new attributes and more methods

##### Multiple inheritance

Sub class can have multiple inheritance there is an order for finding variable and method

- mro(): method resolution order

```python
class A:
    def hello(self):
        print("hello, A")
class B(A):
    def hello(self):
        print("hello, B")
class C(A):
    def hello(self):
        print("Hello, C")
class D(B,C):
    pass

x = D()
x.hello()
D.mro()
```

```python
class Animal:
    def __init__(self,age,name):
        self.age = age
        self.name = name
    def introduce(self):
        return "animal:" + str(self.name)+":"+str(self.age)

class Cat(Animal):
    def speak(self):
        print("meow")
    def introduce(self):
        return "cat:"+str(self.name)+":"+str(self.age)
    
cat = Cat(3, "cheese")
print(cat.introduce())
cat.speak()
```

- There is a method that can get propertys in superclass: "super().__method__()"

```python
class Person:
    total_count = 0
    def __init__(self, name, age):
        self.name = name
        self.age = age
        Person.total_count += 1
    def introduce(self):
        print(f"name:{self.name}")

class Student(Person):
    def __init__(self, name, age):
        super(Student, self).__init__(name, age)
    def introduce(self):
        print(f"Student.age:{self.age}")
    def display_count(self):
        print(f"count:{self.total_count}")

std = Student("Teddy", 20)
std.introduce()
std.display_count()
```

### Practice
Create, inherit, and test a class that matches the geometry below.
( ùúã=3.14)

```python
class Shape:
    def __init__(self,c):
        self.color = c

class Rectangle(Shape):
    def __init__(self, c, h, w):
        super().__init__(c)
        self.height = h
        self.width = w
    def getArea(self):
        return self.height*self.width
    
class Triangle(Shape):
    def __init__(self, c, b, h):
        super().__init__(c)
        self.base = b
        self.height = h
    def getArea(self):
        return self.base*self.height/2
    
class Circle(Shape):
    def __init__(self, c, r):
        super().__init__(c)
        self.radius = r
    def getArea(self):
        return self.radius**2*3.14
```

```python
s1 = Rectangle('black', 2, 5)
print(s1.getArea())

s2 = Triangle('blue', 2, 5)
print(s2.getArea())

s3 = Circle('red', 5)
print(s3.getArea())
```
