### Stack / Queue + Deque
stack and queue is a kind of Liner data structure
- Data is arranged in linear sequence
- Data items can be traversed through sequentially
- Data structure elements are accessed or placed in contiguous memory location (together in sequence)


#### Stack

- Last In First Out(LIFO) linear data structure
- accessible at only one end of the sequence



- It is used for Undo functions 

#### Queue


- enqueue(): It inserts information at the rear of the Queue(efficent)
- dequeue(): It removes information at the front of the Queue(unefficent)
- Useage: call center, Printer, waiting line
